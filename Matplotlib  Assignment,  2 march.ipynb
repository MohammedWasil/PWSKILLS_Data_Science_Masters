##### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.
##### Ans- Scatter plot is a chart that shows the relationship between the two variables.They are  an incredibly powerful chart type alllowing viewers to  immediately understand  a relationship which would to see in almost any other form.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("This is x label data")
plt.ylabel("This is y label data")
```

##### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

##### Ans- we use subplot because it creates a figure and a grid of subplots with a single call while providing  reasonable control over how the individual plots are created.

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x,y,marker='x',color='r')
plt.subplot()
```

##### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
##### Ans-  A bar plot is graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.
##### Bar plot is used when we are showing segments of information.


```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel('company',size=20)
plt.ylabel('profit',size=20)
```

```python
plt.barh(company,profit)
plt.xlabel('company',size=20)
plt.ylabel('profit',size=20)
```
